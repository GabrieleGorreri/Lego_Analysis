  ### ***datasets/lego_sets.csv***
  * **set_num**: A code that is unique to each set in the dataset. This column is critical, and a missing value indicates the set is a duplicate or invalid!

  * **theme_name**: The name of the sub-theme of the set.
  * **parent_theme**: The name of the parent theme the set belongs to. Matches the `name` column of the `parent_themes` csv file.

  ### ***datasets/parent_themes.csv***

  * **id**: A code that is unique to every theme.
  name: The name of the parent theme.
  * **is_licensed**: A Boolean column specifying whether the theme is a licensed theme.

```python
# Import libraries and file

import pandas as pd

df = pd.read_csv('lego_sets.csv')

df.head()
```

```python
theme = pd.read_csv('parent_themes.csv')

theme.head(15)
```

# Question 1: What percentage of all licensed sets ever released were Star Wars themed?

**Remove NAN/duplicates values**

```python
df.shape
```

```python
# Numbers of NaN rows

df[df['set_num'].isna()]
```

```python
# Remove NaN rows...we expect 11986-153 = 11833 rows

df = df[df['set_num'].notna()]
df.shape
```

**Merge 2 datasets**

```python
merged = df.merge(theme, left_on='parent_theme', right_on='name')
merged.drop(columns='name_y', inplace=True)
merged.head()
```

```python
merged.shape
```

**Create new column**

```python
merged['parent_theme'].unique()
```

```python
# WHERE CONDITIONS

import numpy as np

merged['flag_star_wars'] = np.where((merged['theme_name']=='Star Wars') | (merged['parent_theme']=='Star Wars'), 1, 0)
merged.head()
```

```python
# Check

merged[merged['flag_star_wars'] == 1]
```

**Percentage**

```python
the_force = len(merged[(merged['flag_star_wars'] == 1) & (merged['is_licensed'] == True)]) / len(merged['is_licensed'] == True)
the_force
```

# Question 2: In which year was Star Wars not the most popular licensed theme (in terms of number of sets released that year)?

```python
# Licensed dataframe

licensed = merged[merged['is_licensed']] # This column is a booleand, and that means I cannot use ==True
licensed.head()
```

```python
licensed_sorted = licensed.sort_values('year')
licensed_sorted['count'] = 1


licensed_sorted.groupby(['year','parent_theme']).sum()
```

```python
summed_df = licensed_sorted.groupby(['year','parent_theme']).sum().reset_index()
summed_df
```

In this case we can sort the df and its value per year, and eliminate duplicates with a lesser value

```python
# Drop_duplicates has keep=first as default

max_df = summed_df.sort_values('count', ascending=False).drop_duplicates(['year'])

max_df.sort_values('year', ascending = False, inplace=True)
max_df[['year',	'parent_theme', 'num_parts', 'id','is_licensed',	'flag_star_wars',	'count']]
```

# Question 3: Break down number of sets by year

```python
clean_df = merged[~merged['set_num'].isnull()]
clean_df['count'] = 1
sets_per_year = clean_df.groupby(['year']).sum().reset_index()[['year','count']]

sets_per_year
```

```python
for index,row in sets_per_year.iterrows():
  print(row['year'],row['count'])
```
